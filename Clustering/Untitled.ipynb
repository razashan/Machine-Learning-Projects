# K-Means Clustering

Clustering is an unsupervised learning style that can provide some insights. Clustering algorithms attempt to group or co-locate samples that are similar or close to each other.

```python
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
iris= datasets.load_iris()
```

```python
print(iris.DESCR)
```

```python
iris.target_names
```

```python
len(iris.data)
```

```python
X = iris.data[:,:2]
y = iris.target

plt.scatter(X[:,0],X[:,1],c =y)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
```

```python
km = KMeans(n_clusters = 3,n_jobs = 4, random_state = 21)
km.fit(X)
```

```python
km_labels = km.labels_
fig,axes = plt.subplots(1,2,figsize=(12,5))
axes[0].scatter(X[:,0],X[:,1],c = km_labels)
axes[1].scatter(X[:,0],X[:,1],c = y)
axes[0].set_xlabel('Sepal length')
axes[0].set_ylabel('Sepal width')
axes[1].set_xlabel('Sepal length')
axes[1].set_ylabel('Sepal width')
axes[0].set_title('Predicted')
axes[1].set_title('Actual')
```
